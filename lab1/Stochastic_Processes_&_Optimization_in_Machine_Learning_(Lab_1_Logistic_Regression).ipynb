```
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
```

```
# load csv file
df  = pd.read_csv('/content/drive/MyDrive/NTUA FINANCIAL ENGINEERING MASTER/2nd Semester/Stochastic Processes & Optimization in Machine Learning/Lab1/SMSSpamCollection.csv', sep='\t', header=None, names=['label', 'message'])
print(df)
```

```
# split dataset
X_train_raw, X_test_raw, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)

# vectorize data
vectorizer = TfidfVectorizer()
Xtrain = vectorizer.fit_transform(X_train_raw)
Xtest = vectorizer.transform(X_test_raw)

# fit Logistic Regression model
model = LogisticRegression()
model.fit(Xtrain, y_train)

# calculate accuracy based on test set
accuracy = model.score(Xtest, y_test)
print(f"Model accuracy: {accuracy:.4f}")
```

For what reasons do you think that applying the linear regression method is not suitable for this particular problem?

Linear Regression is not suitable for SMS spam classification because it predicts continuous numeric values rather than discrete categories like 'ham' or 'spam', and can produce outputs below 0 or above 1, which are meaningless as probabilities. Unlike Logistic Regression, which uses the sigmoid function to constrain outputs between 0 and 1, Linear Regression is also highly sensitive to outliers that can distort the decision boundary, and its core assumptions of linearity and constant error variance are violated when the target variable is binary.
